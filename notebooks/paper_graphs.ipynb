# Paper Graphs

This notebook loads the latest ultrasupertest structural run for each model and renders one vertical graph stack per model.

Coverage:
- top-k target-layer SVD profile
- Signal A and Signal B
- spectral gap, top1 energy, stable/effective rank ratio
- PCS next jump, next curvature, cross scores, cross curvature
- norm CV, spectral entropy, row alignment
- attention aggregate and attention contrast graphs

All plots are single-column; nothing is laid out side by side.

```python
from pathlib import Path
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set_theme(style="whitegrid", context="talk")
plt.rcParams.update(
    {
        "figure.figsize": (18, 4.5),
        "axes.grid": True,
        "grid.alpha": 0.25,
        "axes.spines.top": False,
        "axes.spines.right": False,
    }
)

ROOT = Path.cwd().resolve()
for p in [ROOT, *ROOT.parents]:
    if (p / "notebooks" / "new-gen" / "paper_graphs_support.py").exists():
        ROOT = p
        break

sys.path.insert(0, str(ROOT / "notebooks" / "new-gen"))

from paper_graphs_support import load_run_manifest, render_paper_graphs
```

```python
MODEL_FILTER = None  # e.g. ['gpt2-xl', 'mistral-7b-v0.1']
RUN_SELECTION = "latest_model"
MAX_FILES = 500
INCLUDE_BASELINE = True
SHOW_STD_BAND = True
TOPK_SVD_RANKS = 30
SAVE_FIGURES = False
FIGURE_OUTPUT_DIR = ROOT / "notebooks" / "new-gen" / "paper_graphs_output"

summary_df, plot_runs, skipped_runs = load_run_manifest(
    root=ROOT,
    model_filter=MODEL_FILTER,
    run_selection=RUN_SELECTION,
    max_files=MAX_FILES,
)

display(summary_df)
if skipped_runs:
    print("Skipping models with no valid tests:", ", ".join(run["model"] for run in skipped_runs))
print(f"Renderable models: {len(plot_runs)}")
```

```python
render_paper_graphs(
    plot_runs,
    include_baseline=INCLUDE_BASELINE,
    show_std_band=SHOW_STD_BAND,
    topk_svd_ranks=TOPK_SVD_RANKS,
    save_figures=SAVE_FIGURES,
    output_dir=FIGURE_OUTPUT_DIR,
)
```
