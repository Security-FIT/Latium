# Prefixtest Visualizer

Loads the latest prefix experiment artifact, renders grouped spectral plots, composite-style signal plots, summary tables, and the paper-style graph stack.

```python
from pathlib import Path
import importlib
import sys

cwd = Path.cwd().resolve()
prefixtest_dir = cwd if (cwd / 'prefixtest_support.py').exists() else cwd / 'prefixtest'
if str(prefixtest_dir) not in sys.path:
    sys.path.insert(0, str(prefixtest_dir))

import prefixtest_support as _prefixtest_support
_prefixtest_support = importlib.reload(_prefixtest_support)
find_artifact = _prefixtest_support.find_artifact
generate_prefixtest_outputs = _prefixtest_support.generate_prefixtest_outputs

OUTPUT_DIR = prefixtest_dir / 'output'
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
MODEL_FILTER = 'Qwen_Qwen3-8B'

artifact = find_artifact(root=cwd, model_filter=MODEL_FILTER)
outputs = generate_prefixtest_outputs(artifact, OUTPUT_DIR, topk_svd_ranks=30)
metadata = outputs['metadata']
df = outputs['df']
layer_df = outputs['layer_df']

print(f'Using artifact: {artifact}')
print(f"Model: {metadata.get('model')}")
print(f"Case idx: {metadata.get('case_idx')} | Case id: {metadata.get('case_id')}")
print(f"Subject: {metadata.get('subject')}")
print(f"Edited layer: {outputs['edited_layer']}")
print(f"Runs: {len(df)} non-baseline + {int((-1 in set(layer_df.run_index)))} unedited baseline")
```

```python
print('Grouped spectral plots:')
for mode, path in outputs['grouped_curve_paths'].items():
    print(f'  {mode}: {path}')

print('Composite signal plots:')
for mode, path in outputs['composite_signal_paths'].items():
    print(f'  {mode}: {path}')

print(f"Paper stack: {outputs['paper_style_path']}")
print(f"Composite accuracy: {outputs['composite_accuracy_path']}")
print(f"Composite method breakdown: {outputs['composite_method_path']}")
print(f"Summary CSV: {outputs['summary_path']}")
print(f"Mode summary CSV: {outputs['mode_summary_path']}")
```
